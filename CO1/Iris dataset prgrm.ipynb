# program4

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("iris.csv")
plt.figure(figsize=(10,3))
x=data["SepalLengthCm"]
plt.hist(x,bins=30,color="r")
plt.xlabel("SepalLength_Cm")
plt.ylabel("Count")
plt.title("SepalLength")
plt.show()
```
