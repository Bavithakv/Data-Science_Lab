```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("car.csv")
x=data["selling_price"]
plt.hist(x,bins=20,color="blue")
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.title("selling_price")
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
iris=pd.read_csv("car.csv")
plt.plot(iris["km_driven"],"r--")
plt.show
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
iris.plot(kind="scatter",x="selling_price",y="km_driven")
plt.grid()
plt.show()
```
